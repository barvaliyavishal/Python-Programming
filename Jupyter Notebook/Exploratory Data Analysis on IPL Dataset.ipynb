```python
# Importing the required libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Extracting the ipl matches dataset from CSV
ipl = pd.read_csv("matches.csv")
```

```python
ipl.head()
```

```python
#looking at the number of rows and columns in the datase
ipl.shape
```

```python
# get the top 5 players with most man of the match awards
ipl['player_of_match'].value_counts()[0:5]
```

```python
# get the top 5 players with most man of the match awards
ipl['player_of_match'].value_counts()[0:5].keys()
```

```python
# get the top 5 players with most man of the match awards
ipl['player_of_match'].value_counts()[0:5].values
```

```python
plt.figure(figsize=(20,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]))
plt.title("Top 5 man of the match data")
plt.xlabel('Player')
plt.ylabel('Number of awards')
```

```python
# getting the frequency of result column
ipl['result'].value_counts()
```

```python
# find out number of toss winner
ipl['toss_winner'].value_counts()
```

```python
# extracting the records where winner team's batting was first
batting_first = ipl[ipl['win_by_runs']!=0]
batting_first.head()
```

```python
#making a histogram of win_by_runs
plt.figure(figsize=(5,5))
plt.hist(batting_first['win_by_runs'])
plt.title("Destribution of runs")
plt.xlabel("runs")
plt.ylabel("Number of matches")
plt.show()
```

```python
# getting frequency of winners in first batting scenario
batting_first['winner'].value_counts()
```

```python
# Making of barchart for top 5 winner teams with first batting
plt.figure(figsize=(12,6))
plt.bar(list(batting_first['winner'].value_counts()[0:5].keys()),list(batting_first['winner'].value_counts()[0:5]),color=['blue','yellow','orange','purple','green'])
```

```python
# Making of pie chart for top winner teams with first batting
plt.figure(figsize=(12,12))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
```

```python
# looking at the number of matches played in each season
season_match = ipl['season'].value_counts()
```

```python
# making a chart for number of matches played in  each year
plt.figure(figsize=(7,7))
plt.plot(season_match.keys(),season_match.values)
```
