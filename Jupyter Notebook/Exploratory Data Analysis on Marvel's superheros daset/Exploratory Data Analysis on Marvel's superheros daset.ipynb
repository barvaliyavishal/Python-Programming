```python
#loading required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
marvel = pd.read_csv("/home/vishal/C/Dataset/Marvel_dataset/charcters_stats.csv")
```

```python
# have a glance at dataset
marvel.head()
```

```python
# Number of rows and columns in dataser
marvel.shape
```

```python
# getting the number of good bad and neutral heros
marvel['Alignment'].value_counts()
```

```python
# extracting all superheros with good alignment
good = marvel[marvel['Alignment']=='good']
good.head()
```

```python
# getting top 10 superheros with good alignment and highest speed
good.sort_values(by=['Speed'],ascending=False)[0:10]
```

```python
# extracting top 5 superheros with good alignment and highest power
good_power = good.sort_values(by=['Power'],ascending=False)[0:5]
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(good_power['Name'][0:5]),list(good_power['Total'])[0:5],color='r')
```

```python
# superheros with power 100
good[good['Power']==100].shape[0]
```

```python
# extracting all supervillains
bad = marvel[marvel['Alignment']=='bad']
```

```python
# find top5 bad super villains with highest power
bad.sort_values(by=['Power'],ascending=False)[0:5]
```

```python
# find top5 bad super heros with highest Speed
bad.sort_values(by=['Speed'],ascending=False)[0:5]
```

```python
#making a histogram for speed for good superheros
plt.figure(figsize=(10,5))
plt.hist(good['Speed'])
plt.title('Distribution of Speed')
plt.xlabel('Speed')
plt.show()
```
