```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
y=2*x
y2=4*x
```

```python
plt.plot(x,y,color="g",linestyle=':')
plt.plot(x,y2,color="r",linestyle='-')
plt.title("Two Axis x and y")
plt.xlabel("this is x")
plt.ylabel("this is y")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(x,y2)
```

#  Bar Graph

```python
student = {'Bob':90,'Jack':50,'Julia':60}
names = student.keys()
value =student.values()
plt.bar(names,value)
plt.show()
```

```python
student = {'Bob':90,'Jack':50,'Julia':60}
names = student.keys()
value = student.values()
plt.bar(names,value,color='g')
plt.grid(True)
plt.show()
```

## Scatter graph

```python
plt.scatter(names,value)
```

```python
plt.scatter(names,value,marker='*',color='r')
plt.scatter(names,[28,23,49],marker='*')
```

#  Pie chart

```python
student = {'Bob':90,'Jack':50,'Julia':60}
names = student.keys()
value = student.values()
```

```python
plt.pie(value,labels=names,autopct='%0.1f%%')
```

```python
student = {'Bob':90,'Jack':50,'Julia':60}
names = student.keys()
value = student.values()
```

```python
plt.pie(value,labels=names,autopct='%0.1f%%',radius=2)
plt.pie([1],colors=['w'],radius=1)
```
